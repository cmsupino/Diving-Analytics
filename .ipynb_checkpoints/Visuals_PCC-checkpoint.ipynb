```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

```python
#Load Data Frame for analysis 
df_final = pd.read_csv("df_merged_scaled.csv")
```

```python
#Linear regression: ScaledPoints and  PCC
X = df_final[['ScaledPoints']] #Independent variable (official points)
y = df_final['PCC'] #Dependent variable (PCC)

#Train linear regression model
model = LinearRegression()
model.fit(X, y)

#Calculate correlation coefficient (r-value)
r_value, _ = pearsonr(df_final['ScaledPoints'], y)
r_squared = r_value**2

#Predict PCC values across range of official points
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_pred = model.predict(x_range)

#Plot Elements
plt.figure(figsize=(8,6))
plt.scatter(X, y, alpha=0.5, label="Data")
plt.plot(x_range, y_pred, linewidth=3, label=f"Regression Line (r = {r_value:.3f})")
plt.xlabel("ScaledPoints")
plt.ylabel("PCC")
plt.title("Linear Regression: Scaled Points vs PCC")
plt.legend()

#Output
plt.show()
```

```python
#Scatter plot of PCC vs Official Points for all dives and divers
plt.figure(figsize=(10,7))

sns.scatterplot(
    data=df_final,
    x='PCC',
    y='ScaledPoints',    #Use scaled official points
    hue='DiveName',      #Color by dive
    style='LastName',    #Shape by diver
    palette='tab10',
    s=100
)

#Plot Elements
plt.title("PCC vs Official Points (All Dives & Divers)")
plt.xlabel("PCC Score (1-10)")
plt.ylabel("Official Points (scaled 1-10)")
plt.grid(True)
plt.legend(title='Dive / Diver', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

#Calculate correlations per dive and per diver
for dive in df_final["DiveName"].unique():
    df_dive = df_final[df_final["DiveName"] == dive]
    overall_corr = df_dive["PCC"].corr(df_dive["ScaledPoints"])
    print(f"Dive {dive} - Overall correlation: {overall_corr:.3f}")
    
    for diver in df_dive["LastName"].unique():
        df_diver = df_dive[df_dive["LastName"] == diver]
        if len(df_diver) > 1:
            corr = df_diver["PCC"].corr(df_diver["ScaledPoints"])

            #Print Correlations
            print(f"Dive {dive} - {diver} correlation: {corr:.3f}")
```

```python
#Individual Scatter plot for each dive type.

#Loop through each dive
for dive in df_final["DiveName"].unique():
    # Subset for this dive
    df_dive = df_final[df_final["DiveName"] == dive]
    
    #Scatter plot
    plt.figure(figsize=(8,6))
    sns.scatterplot(
        data=df_dive,
        x='PCC',
        y='ScaledPoints',  #Use scaled official points
        hue='LastName',    #Color by diver
        palette='tab10',
        s=100
    )

    #Plot elements 
    plt.title(f"PCC vs Official Points for Dive {dive}")
    plt.xlabel("PCC Score (1-10)")
    plt.ylabel("Official Points (scaled 1-10)")
    plt.grid(True)
    plt.legend(title='Diver')

    #Output
    plt.show()
    
    #Overall correlation for this dive
    overall_corr = df_dive["PCC"].corr(df_dive["ScaledPoints"])
    print(f"Dive {dive} - Overall correlation: {overall_corr:.3f}\n")
```

```python
#Scatter plots of PCC vs Official Points for each diver
divers = df_final["LastName"].unique()
plt.figure(figsize=(14, 10))

#Loop through divers
for i, diver in enumerate(divers, 1):
    df_diver = df_final[df_final["LastName"] == diver]
    
    #Scatter plot
    plt.subplot(3, 2, i)
    sns.scatterplot(
        data=df_diver,
        x="PCC_scaled",
        y="ScaledPoints", 
        hue="DiveName",
        palette="tab10",
        s=100
    )

    #Plot elements
    plt.title(f"PCC vs Official Points - {diver}")
    plt.xlabel("PCC Score (1-10)")
    plt.ylabel("Official Points (scaled 1-10)")
    plt.grid(True)
    plt.legend(title='Dive', bbox_to_anchor=(1.05, 1), loc='upper left')

#Output
plt.tight_layout()
plt.show()

#Print correlations for each diver
for diver in divers:
    df_diver = df_final[df_final["LastName"] == diver]
    corr = df_diver["PCC_scaled"].corr(df_diver["ScaledPoints"])
    print(f"{diver} correlation: {corr:.3f}")
```

```python
#Creation of 2 boxplots exploring the realtionship between PCC and Dive Performed & between PCC and Diver

#Rescale PCC to 1-10 if not already scaled
if 'PCC_scaled' not in df_final.columns:
    pcc_min, pcc_max = df_final["PCC"].min(), df_final["PCC"].max()
    df_final["PCC_scaled"] = 1 + 9 * (df_final["PCC"] - pcc_min) / (pcc_max - pcc_min)

#BOXPLOT PER DIVE:
plt.figure(figsize=(14,6))
#Create Boxplot
sns.boxplot(
    data=df_final,
    x="DiveName",
    y="PCC_scaled",
    palette="Set2",
    whis=[0,100]
)

#Create Striplot
sns.stripplot(
    data=df_final,
    x="DiveName",
    y="PCC_scaled",
    hue="LastName",
    dodge=False,
    marker="o",
    alpha=0.8,
    palette="tab10",
    size=8
)

#Plot Elements
plt.title("Distribution of PCC Scores per Dive")
plt.xlabel("Dive Name")
plt.ylabel("PCC Score (scaled 1-10)")
plt.grid(True, axis='y')
plt.legend(title='Diver', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()


plt.show()

#BOXPLOT PER DIVER:
plt.figure(figsize=(14,6))

#Create BoxPlot
sns.boxplot(
    data=df_final,
    x="LastName",
    y="PCC_scaled",
    palette="Set3"
)

#Create StripPlot 
sns.stripplot(
    data=df_final,
    x="LastName",
    y="PCC_scaled",
    hue="DiveName",
    dodge=True,
    marker="o",
    alpha=0.8,
    palette="tab10",
    size=8
)

#Plot elements
plt.title("PCC Scores per Diver Across Their 5 Dives (Scaled 1-10)")
plt.xlabel("Diver")
plt.ylabel("PCC Score (1-10)")
plt.grid(True, axis='y')
plt.legend(title='Dive Name', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()

#Output
plt.show()
```

```python
#Plot of differences in PCC scores vs Official Scores(Scaled) to show underscoring in divers dive specific scores. 

#Compute underscoring
df_final['Underscore'] = df_final['ScaledPoints'] - df_final['PCC_scaled']

#Average underscoring per diver 
avg_underscore = df_final.groupby('LastName')['Underscore'].mean().sort_values(ascending=False)

#Output
print("\nAverage underscoring per diver (higher = more underscored):")
print(avg_underscore)

#Plot per-dive underscoring
plt.figure(figsize=(10,6))

#Bar plot
sns.barplot(
    data=df_final.sort_values('Underscore', ascending=False),
    x='DiveName',
    y='Underscore',
    hue='LastName',
    palette='tab10'
)

#Plot elements
plt.title("Dive-by-Dive Underscoring per Diver")
plt.xlabel("Dive")
plt.ylabel("Official Points (scaled) - PCC (scaled)")
plt.grid(True, axis='y')
plt.legend(title='Diver', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()

#Ouptut
plt.show()
```

```python
#Total PCC and official points per diver
total_pcc = df_final.groupby('LastName')['PCC_scaled'].sum()
total_official = df_final.groupby('LastName')['ScaledPoints'].sum()

#Ranking difference between PPC and Official Points(Entire Data Frame)
ranking_df = pd.DataFrame({
    'TotalPCC': total_pcc,
    'TotalOfficialPoints': total_official
})
ranking_df['PCC_Rank'] = ranking_df['TotalPCC'].rank(ascending=False)
ranking_df['Official_Rank'] = ranking_df['TotalOfficialPoints'].rank(ascending=False)

#Output
print("\nComparison of PCC and Official Points rankings:")
print(ranking_df.sort_values('Official_Rank'))


#Exclude dive 201B
df_no_201B = df_final[df_final['DiveName'] != '201B']
total_pcc_no201B = df_no_201B.groupby('LastName')['PCC_scaled'].sum()
total_official_no201B = df_no_201B.groupby('LastName')['ScaledPoints'].sum()

#Ranking difference between PPC and Official Points(exclude 201B)
ranking_df_no201B = pd.DataFrame({
    'TotalPCC': total_pcc_no201B,
    'TotalOfficialPoints': total_official_no201B
})
ranking_df_no201B['PCC_Rank'] = ranking_df_no201B['TotalPCC'].rank(ascending=False)
ranking_df_no201B['Official_Rank'] = ranking_df_no201B['TotalOfficialPoints'].rank(ascending=False)

#Output
print("\nComparison of PCC and Official Points rankings (excluding 201B):")
print(ranking_df_no201B.sort_values('Official_Rank'))
```

```python
#Per-diver correlations excluding 201B
diver_corrs = {}

#Loop through divers 
for diver in df_no_201B['LastName'].unique():
    df_diver = df_no_201B[df_no_201B['LastName'] == diver]
    diver_corrs[diver] = df_diver['PCC_scaled'].corr(df_diver['ScaledPoints']) if len(df_diver) >= 2 else None

#Overall correlation
overall_corr = df_no_201B['PCC_scaled'].corr(df_no_201B['ScaledPoints'])

#Output
print("Overall PCC vs Official Score correlation (excluding 201B):", overall_corr)

print("Per-diver PCC vs Official Score correlations (excluding 201B):")
for diver, corr in diver_corrs.items():
    print(f"{diver}: {corr:.3f}" if corr is not None else f"{diver}: Not enough dives")
```
